```python
import pytesseract
from PIL import Image
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Folder where your images are located
image_folder = "Project-Folder"

# Collect all image file paths (including subfolders)
image_paths = []
for root, dirs, files in os.walk(image_folder):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_paths.append(os.path.join(root, file))

# Check and display image + improved OCR
if not image_paths:
    print("‚ùå No images found in the folder. Check the folder path and contents.")
else:
    for img_path in image_paths:
        # Load image and show it
        image = Image.open(img_path)
        plt.imshow(image)
        plt.axis('off')
        plt.title(f"Image: {os.path.basename(img_path)}", fontsize=10)
        plt.show()

        # Preprocess image for better OCR (grayscale + adaptive threshold)
        image_cv = np.array(image)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_RGB2GRAY)
        thresh = cv2.adaptiveThreshold(gray, 255,
                                       cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                       cv2.THRESH_BINARY, 11, 2)

        # Convert back to PIL for Tesseract
        preprocessed_image = Image.fromarray(thresh)

        # Extract text using Tesseract (with layout config)
        text = pytesseract.image_to_string(preprocessed_image, config='--psm 6')

        # Show extracted text
        print(f"üìù Extracted Text from: {os.path.basename(img_path)}")
        print(text.strip())
        print("\n" + "="*80 + "\n")
```
