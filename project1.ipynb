```python
# Plot: Price vs Rating by Category
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Price (INR)", y="Rating", hue="Category", s=100)
plt.title("Price vs Rating by Category")
plt.xlabel("Price (INR)")
plt.ylabel("Rating")
plt.grid(True)
plt.show()
```

```python
# Average Price by Brand
avg_price_brand = df.groupby("Brand")["Price (INR)"].mean().sort_values()

plt.figure(figsize=(8, 6))
sns.barplot(x=avg_price_brand.values, y=avg_price_brand.index, palette="viridis")
plt.title("Average Price by Brand")
plt.xlabel("Average Price (INR)")
plt.ylabel("Brand")
plt.grid(True)
plt.show()
```

```python
# Price Distribution by RAM Size
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="RAM (GB)", y="Price (INR)", palette="Set2")
plt.title("Price Distribution by RAM Size")
plt.xlabel("RAM (GB)")
plt.ylabel("Price (INR)")
plt.grid(True)
plt.show()
```

```python
# Count of Products by Category
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="Category", palette="pastel")
plt.title("Number of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
def search_products(keyword):
    results = df[df["Title"].str.contains(keyword, case=False, na=False)]
    if results.empty:
        print(f"No products found for: {keyword}")
    else:
        print(f"\n🔍 Found {len(results)} products for '{keyword}':\n")
        display(results[["Title", "Brand", "Price (INR)", "Rating", "Category"]])
```

```python
search_products("asus")
```

```python
search_products("gaming")
search_products("MacBook")
search_products("HP")
```

```python
def search_by_brand(brand):
    filtered = df[df["Brand"].str.lower() == brand.lower()]
    display(filtered[["Title", "Brand", "Price (INR)", "Rating", "Category"]])
```

```python
search_by_brand("Apple")
```

```python
def search_by_category_and_price(category, min_price, max_price):
    filtered = df[
        (df["Category"].str.lower() == category.lower()) &
        (df["Price (INR)"] >= min_price) &
        (df["Price (INR)"] <= max_price)
    ]
    display(filtered[["Title", "Brand", "Price (INR)", "Rating", "Category"]])
```

```python
search_by_category_and_price("Gaming", 50000, 90000)
```
